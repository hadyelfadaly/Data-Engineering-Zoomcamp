```python
df = pd.read_parquet('green_tripdata_2025-11.parquet')

df.head()
```

```python
df.head(n=0).to_sql(name='greenData_NOV25', con=engine, if_exists='replace') 
# `head(n=0)` makes sure we only create the table, we don't add any data yet.

df.to_sql(name='greenData_NOV25', con=engine, if_exists='append') #inserting data
```

```python
df_lookup.head(n=0).to_sql(name='lookupData', con=engine, if_exists='replace') 
# `head(n=0)` makes sure we only create the table, we don't add any data yet.

df_lookup.to_sql(name='lookupData', con=engine, if_exists='append') #inserting data
```
